## ¿Qué aerolíneas ejecutan más vuelos?

```python
%%sql SELECT count(*) as value, Description
    FROM ontime
    INNER JOIN airline_id
    ON airline_id.code = ontime.DOT_ID_Reporting_Airline
    GROUP BY DOT_ID_Reporting_Airline
    ORDER BY value DESC
```

## ¿Cuales aerolíneas viajan más entre semana?

```python
%%sql SELECT count(*) as value, Description
    FROM ontime
    INNER JOIN airline_id
    ON airline_id.code = ontime.DOT_ID_Reporting_Airline
    WHERE DayOfWeek < 6
    GROUP BY DOT_ID_Reporting_Airline
    ORDER BY value DESC
    
```

## ¿Cuales aerolíneas viajan más fines de semana?

```python
%%sql SELECT count(*) as value, Description
    FROM ontime
    INNER JOIN airline_id
    ON airline_id.code = ontime.DOT_ID_Reporting_Airline
    WHERE DayOfWeek > 5
    GROUP BY DOT_ID_Reporting_Airline
    ORDER BY value DESC
```

## ¿Cual de las aerolíneas han tenido mayor proporcion de cancelación?

```python
%%sql SELECT count(*)
as value, Description
FROM ontime
INNER JOIN airline_id
ON airline_id.code = ontime.DOT_ID_Reporting_Airline
WHERE cancelled = 1.0
GROUP BY DOT_ID_Reporting_Airline
ORDER BY value DESC
```

## ¿Qué aerolínea es menos frecuente en retrasos de salida?

```python
%%sql SELECT count(*) * 100 / (SELECT count(*) FROM ontime GROUP BY DOT_ID_Reporting_Airline)
as value, Description
FROM ontime
INNER JOIN airline_id
ON airline_id.code = ontime.DOT_ID_Reporting_Airline
WHERE depdelay > 0
GROUP BY DOT_ID_Reporting_Airline
ORDER BY value ASC
```

## ¿Que aerolínea llega puntual a su destino mas frecuentemente?

```python
%%sql SELECT count(*) * 100 / (SELECT count(*) FROM ontime GROUP BY DOT_ID_Reporting_Airline)
as value, Description
FROM ontime
INNER JOIN airline_id
ON airline_id.code = ontime.DOT_ID_Reporting_Airline
WHERE arrdelay <= 0
GROUP BY DOT_ID_Reporting_Airline
ORDER BY value DESC
```

## ¿Cuál aerolínea presentan más tiempo de retraso al despegue promedio?

```python
%%sql SELECT AVG(depdelay)
as value, Description
FROM ontime
INNER JOIN airline_id
ON airline_id.code = ontime.DOT_ID_Reporting_Airline
GROUP BY DOT_ID_Reporting_Airline
ORDER BY value ASC
```

## ¿Cuál aerolínea presentan más tiempo de retraso promedio para llegar a su destino?

```python
%%sql SELECT AVG(arrdelay)
as value, Description
FROM ontime
INNER JOIN airline_id
ON airline_id.code = ontime.DOT_ID_Reporting_Airline
GROUP BY DOT_ID_Reporting_Airline
ORDER BY value ASC
```

## ¿Cuál es la causa principal de atrasos para los vuelos entre las aerolíneas?

```python
%%sql SELECT airline_id.Description AS Aerolinia, cancellation.Description AS CausaCancelacion
FROM ontime
INNER JOIN airline_id
ON airline_id.code = ontime.DOT_ID_Reporting_Airline
INNER JOIN cancellation
ON cancellation.code = ontime.CancellationCode
WHERE CancellationCode != ""
GROUP BY DOT_ID_Reporting_Airline
```

## ¿Que aerolinia pasa mas tiempo en vuelo en promedio?

```python
%%sql SELECT avg(AirTime) as TiempoEnAire, Description
FROM ontime
INNER JOIN airline_id
ON airline_id.code = ontime.DOT_ID_Reporting_Airline
GROUP BY DOT_ID_Reporting_Airline
ORDER BY TiempoEnAire ASC
```

## ¿Qué aerolínea cubre mayor cantidad de destinos?

```python
%%sql SELECT count(DISTINCT dest) as NumeroDestinos, Description
FROM ontime
INNER JOIN airline_id
ON airline_id.code = ontime.DOT_ID_Reporting_Airline
GROUP BY DOT_ID_Reporting_Airline
ORDER BY NumeroDestinos DESC
```
