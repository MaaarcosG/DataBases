¿Cuántas aerolíneas viajan más entre semana?

```python
%%sql SELECT count(*) as value, Description
    FROM ontime
    INNER JOIN airline_id
    ON airline_id.code = ontime.DOT_ID_Reporting_Airline
    WHERE DayOfWeek < 6
    GROUP BY DOT_ID_Reporting_Airline
    ORDER BY value DESC
    
```

¿Cual de las aerolíneas han tenido mayor cancelación entre los últimos años?

```python
%%sql SELECT count(*) as value, Description
    FROM ontime
    INNER JOIN airline_id
    ON airline_id.code = ontime.DOT_ID_Reporting_Airline
    WHERE Cancelled = 1.0
    GROUP BY DOT_ID_Reporting_Airline
    ORDER BY value DESC
```
